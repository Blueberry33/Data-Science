```python
#importing required libraries
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#linking the data file to the program
file_link= "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
scores=pd.read_csv(file_link)
x=scores.iloc[:,0].values
y=scores.iloc[:,1].values
x=x.reshape(-1,1)
y=y.reshape(-1,1)

#For displaying the table
scores.head()
```

```python
#printing hours spent by students
print(x)
```

```python
#printing Their scores
print(y)
```

```python
#plotting the parameters
scores.plot()
```

```python
#scatter plot of the given table
scores.plot(kind="scatter", x="Hours", y="Scores", label="Time vs Scores")
```

```python
#getting test data from the existing data set
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#implementing linear regression
from sklearn.linear_model import LinearRegression
output=LinearRegression()
output.fit(xtrain,ytrain)
```

```python
#prediction
ypred=output.predict(xtest)
ypred=ypred.reshape(-1,1)
ypred
```

```python
#Comparison between original value and predicted value
predicted=pd.DataFrame({'Original value':ytest.flatten(),'Predicted Value':ypred.flatten()})
predicted
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
mean_absolute_error(ytest,ypred),
mean_squared_error(ytest,ypred),
r2_score(ytest,ypred)
```

```python
#Visualising Linear Regression
line= output.coef_*x+output.intercept_
pt.title('linear Regression')
pt.xlabel('Hours spent')
pt.ylabel('Scores Obtained')
pt.scatter(x,y)
pt.plot(x,line,color="red")
pt.show()
```

```python
#solution of the given problem statement
ypred1=output.predict([[9.25]])
print("Score for 9.25 hours is: ",ypred1)
```
